```python
#!/usr/bin/env python
import urllib.request
from bs4 import BeautifulSoup

min_year = 3000
max_year = 1000
dire_set = set([])
star_set = set([])
pos_set = set([])
type_set = set([])

print(u'豆瓣电影TOP250:\n 序号 \t 影片名\t 导演\t 主演\t 时间\t 地区\t 类型\t 评分\t 评价人数\t 评价')
def crawl(url):
    global min_year
    global max_year
    global dire_set
    global star_set
    global pos_set
    global type_set
    
    headers = {'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US; rv:1.9.1.6) Gecko/20091201 Firefox/3.5.6'}
    req = urllib.request.Request(url, headers=headers)
    page = urllib.request.urlopen(req, timeout=60)
    contents = page.read()
    soup = BeautifulSoup(contents,"lxml")
    for tag in soup.find_all('div', class_='item'):
        try:                  
            m_order = int(tag.find('em', class_='').get_text())
            
            m_name = tag.span.get_text()
            mess = tag.find('p', class_='').get_text().replace(':',' ').split()
            for i in range(len(mess)):
                if mess[i] == '导演':
                    m_director = mess[i+1]
                if mess[i] == '主演':
                    m_starring = mess[i+1]
            
            m_yl = tag.find('p', class_='').get_text().split('\n')[2].split()
            m_year = int(m_yl[0][0:4])
            m_location = m_yl[2]
            
            if m_year > max_year:
                max_year = m_year
            if m_year < min_year:
                min_year = m_year
            
            # get type
            mess = tag.find('p', class_='').get_text()
            type_index = mess.rfind('/') + 1 
            m_type = mess[type_index : len(mess)].split()
            
            # get others
            m_rating_score = float(tag.find('div', class_='star').find('span', class_='rating_num').get_text())
            m_rating_num = (tag.find('div', class_='star').find('span',class_='',content='').get_text())[0:-3]
            m_comments = tag.find("span", class_="inq").get_text()
            
            dire_set.add(m_director)
            star_set.add(m_starring)
            pos_set.add(m_location)
            m_type_str = '['
            for type_ in m_type:
                type_set.add(type_)
                m_type_str += "'"
                m_type_str += type_
                m_type_str += "',"
            m_type_str = m_type_str[0:len(m_type_str)-1] + ']'
            
        except AttributeError:
            print("movie('%s','%s','%s',%s,'%s',%s,%s,%s,'%s')." % (m_name, m_director, m_starring, 
                                                     m_year, m_location, m_type_str, m_rating_score, m_rating_num, "NO COMMENTS"))
        else:
            print("movie('%s','%s','%s',%s,'%s',%s,%s,%s,'%s')." % (m_name, m_director, m_starring, 
                                                     m_year, m_location, m_type_str, m_rating_score, m_rating_num, m_comments))
pagenumber = []
for i in range(10):
    page_number = 25*i
    pagenumber.append(page_number)
pagelist = list(map(str, pagenumber))
 
BASE_URL = 'http://movie.douban.com/top250?start='
LAST_URL = '&filter=&type='
for url in [ BASE_URL + MID_URL + LAST_URL for MID_URL in pagelist ]:
    crawl(url)
    
print(min_year,max_year)
print(len(dire_set),dire_set)
print(len(star_set),star_set)
print(len(pos_set),pos_set)
print(len(type_set),type_set)
```
