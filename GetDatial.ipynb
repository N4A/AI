```python

#!/usr/bin/env python
#coding:utf-8
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup

#get movie's detail infos
def movie_detail(name):
    # get moive through search
    so_url = "http://movie.douban.com/subject_search?search_text="
    headers = {'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US; rv:1.9.1.6) Gecko/20091201 Firefox/3.5.6'}
    req = urllib.request.Request(so_url+urllib.parse.quote(name),headers=headers)
    # GET searh result 0
    data = urllib.request.urlopen(req,timeout=60).read().decode('UTF-8')
    soup = BeautifulSoup(data,'lxml')
    #get movie linl
    movie_link = soup.find('div', class_='pl2').find('a').attrs['href']
    #get movie info page
    ymdata = urllib.request.urlopen(movie_link).read().decode('UTF-8')
    soup = BeautifulSoup(ymdata,'lxml')
    # get movie info
    try:
        infos = soup.find('span',{'property':'v:summary'}).get_text().replace("'","").replace('\n','').replace(' ','')
    except AttributeError:
        print("info('%s','%s')." % (name, "NO INTRODUCTION") )
    else:
        print("info('%s','%s')." % (name, infos) )

#get all movies
def crawl(url):
    headers = {'User-Agent':'Mozilla/5.0 (Windows; U; Windows NT 6.1; en-US; rv:1.9.1.6) Gecko/20091201 Firefox/3.5.6'}
    req = urllib.request.Request(url, headers=headers)
    page = urllib.request.urlopen(req, timeout=60)
    contents = page.read()
    soup = BeautifulSoup(contents,"lxml")
    for tag in soup.find_all('div', class_='item'):                
        m_name = tag.span.get_text()
        movie_detail(m_name)
    

#main processs
pagenumber = []
for i in range(10):
    page_number = 25*i
    pagenumber.append(page_number)
pagelist = list(map(str, pagenumber))
 
BASE_URL = 'http://movie.douban.com/top250?start='
LAST_URL = '&filter=&type='
for url in [ BASE_URL + MID_URL + LAST_URL for MID_URL in pagelist ]:
    crawl(url)
```
